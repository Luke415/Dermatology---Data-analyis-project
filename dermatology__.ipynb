### 1) For each class, find the average value of scaling.

```python
t1 = derm.groupby("class")["scaling"].mean().round(4).nlargest(6).reset_index()
t1 = t1.sort_values("scaling", ascending=False)
t1
```

```python
plt.figure(figsize=(9,4))
sns.barplot( data=t1, x="scaling", y="class")
plt.title("Class by scaling", fontsize=15)
# plt.xlabel('h')
# plt.ylabel("Scaling")
plt.show()
```

### 4) For each class, find:
#### i)	  The mean erythema
#### ii)  The maximum scaling
#### iii) The count of patients

```python
t2 = derm.groupby("class")["erythema"].mean().round(4).reset_index() 
t2 = t2.nlargest(6,"erythema")
print(t2)
```

```python
plt.figure(figsize=(11,5))
ax = sns.barplot(data=t2, x="erythema", y="class", hue="erythema", errorbar='sd')
ax.set_title("Class distribution by erythema", fontsize=14)
ax.set_xlabel("Erythema")
ax.set_ylabel("Class")
plt.show()
```

```python
 t3 = derm.groupby("class")["erythema"].max().reset_index()
```

```python
plt.figure(figsize=(11,5))
ax = sns.boxplot(data=t3, x="erythema", y="class", hue="erythema")
ax.set_title("Maximum count of erythema", fontsize=14)
ax.set_xlabel("Erythema")
ax.set_ylabel("Class")
plt.show()
```

```python
t4 =derm.groupby("class")["age"].count().reset_index()
```

```python
plt.figure(figsize=(11,5))
ax = sns.barplot(data=t4, y="class", x="age", hue="age")
ax.set_title("Class by age", fontsize=14)
ax.set_xlabel("Age", fontsize=10)
ax.set_ylabel("Class", fontsize=10)
plt.show()
```

### 7) Create a new column called age_group where:
#### i)	  Age < 30 → "Young"
#### ii)  Age 30–60 → "Middle"
#### iii) Age > 60 → "Senior"

```python
derm["age_group"] = derm["age"].apply(lambda x: "Young" if x < 30 else ("Middle" if x <= 60 else "Senior"))
```

```python
# New Column
derm["age_group"]
```

#### 8) Find how many patients belong to each age_group.

```python
t5 = derm.groupby("age_group")["age"].count().reset_index()
t5 = t5.sort_values("age", ascending=False)
t5
```

```python
plt.figure(figsize=(12,5))
plt.pie(t5["age"], labels=t5["age_group"],  autopct='%1.1f%%', startangle=60, wedgeprops={"edgecolor":"white"})
plt.title("Counts of Ages by age_group", fontsize=14)
plt.axis("equal")
plt.legend()
plt.show
```
